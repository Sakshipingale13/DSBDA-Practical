```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv("/content/job_placement.csv")
```

```python
data.info()
```

```python
data.shape
```

```python
grouped_data = data.groupby('age')

# Calculate summary statistics for the numeric variable (e.g., income)
summary_stats = grouped_data['salary'].agg(['mean', 'median', 'min', 'max', 'std'])
print(summary_stats)
```
